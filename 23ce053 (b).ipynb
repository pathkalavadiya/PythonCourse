```python
#create lists of many types
#1.
list1=[10,20,30,40]
list2=["charusat","cspit","depstar","cricket"]
list3=["india",10,11.63,"charusat",56]

print(list1)
print(list2)
print(list3)

#2.
#access the given indices
print(list1[0])

#print the silings
print(list1[1:4])

#update the list
list3[3]="cspit"
print(list3)

#3.
#append operation
list1.append(50)
print(list1)

#inseart operation
list3.insert(2,"farmer")
print(list3)

#remove operation
list2.remove("depstar")
print(list2)

#pop operation
list1.pop(1)
print(list1)
list2.pop(1)
print(list2)

#4.concatenate lists 
mergelist=list1+list2
print(mergelist)

#repeate list
list=list1*2
print(list)

#5.
#comprehension
list4=[i**2 for i in range(1,11)]
print(list4)

#filter even number of list
list5=[i for i in list4 if i%2==0]
print(list5)

#6.

#sorting list
list6=[1,5,3,7,2,9]
list6.sort()
print(list6)

#reversing 
list7=[1,5,3,7,2,9]
list7.reverse()
print(list7)

#coping 
list8=list7.copy()
print(list8)    

#7.
#remove duplicates in list
list9=[1,2,2,3,4,4,5,6,6]
list10 = []
[list10.append(i) for i in list9 if i not in list10]
print(list10)
```
